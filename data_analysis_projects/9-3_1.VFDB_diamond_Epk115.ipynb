```python
for path_db_vf in lst_db : 
    
    db1 = path_db_vf.split('/')[-1].split('.dmnd')[0]
    path_out = dir_diamond_result + "%s_%s.out" % (name_epk115, db1)

    df_blast = pd.read_csv( path_out, sep = "\t", header = -1 )
    df_blast.columns = [ "gene_id", "target", "pidentity", "length", "mismatch", "gapopen", "qstart", "qend", "sstart", "send", "evalue", "bitscore" ]

    lst_total_length = []
    lst_coverage = []
    
    lst_gene_id = []
    lst_target = []
    lst_target_info = []
    # protein similarity
    lst_pidentity = []
    lst_length = []
    lst_evalue = []
    lst_bitscore = []
    count = 0
    lst_index = []
    
    for idx, row in df_blast.iterrows() :
        len_gene_id = len(dic_epk115[row['gene_id']])
        coverage = round(row['length'] / float(len_gene_id) * 100, 2)
        
        if row['pidentity'] < 90 :
            continue
        if coverage < 90 :
            continue
        info = dic_vf_name[row['target']]
        count += 1
        lst_index.append(count)
        lst_gene_id.append(row['gene_id'])
        lst_target.append(row['target'])
        lst_target_info.append(info)
        lst_pidentity.append(row['pidentity'])
        lst_length.append(row['length'])
        lst_evalue.append(row['evalue'])
        lst_bitscore.append(row['bitscore'])           
        
        lst_total_length.append(len_gene_id)
        lst_coverage.append(coverage)
                  
    df_edit = pd.DataFrame(lst_gene_id)
    df_edit['target'] = lst_target
    df_edit['target_info'] = lst_target_info
    df_edit['total_length'] = lst_total_length
    df_edit['length'] = lst_length
    df_edit['coverage'] = lst_coverage
    df_edit['pidentity'] = lst_pidentity
    df_edit['evalue'] = lst_evalue
    df_edit['bitscore'] = lst_bitscore
    
    df_edit.columns = [ "EpK1/15", "VFDB", "VFDB_info", "EpK1/15_length", "aligned_length", "coverage", "pidentity", "evalue", "bitscore" ]
    df_edit.index = lst_index
    
    tsv_df_edit = dir_notebook + "9-3.VFDB/result_blastp_coverage.pidentity_more_90p_%s.tsv" %db1
    
    df_edit.to_csv(tsv_df_edit, sep = '\t')
    print df_edit
    
```
